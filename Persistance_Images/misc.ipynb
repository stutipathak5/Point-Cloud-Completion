```python
from itertools import product

import time
import numpy as np
from sklearn import datasets
from scipy.stats import multivariate_normal as mvn
import matplotlib.pyplot as plt

from ripser import Rips
from persim import PersistenceImager
import open3d as o3d
import numpy as np
import h5py
import os
import random

def list_subfolders(folder_path):
    subfolders = [f.path for f in os.scandir(folder_path) if f.is_dir()]
    return subfolders

def load(folder_path):
    subfolders = list_subfolders(folder_path)
    a=[]
    for i in subfolders:
        files = [f.path for f in os.scandir(i) if f.is_file()]
        for j in files:
            a.append(j)
    return a


complete = load(r"\\datanasop3mech\ProjectData\3_phd\Stuti\PCC&PSS\Data\ShapeNetCompletion\train\complete")
print(len(complete))
print(complete[:5])
random.shuffle(complete)
print(complete[:5])
```

```python
s0 = 0
s1 = 0
for i in complete[:100]:
    pcd = o3d.io.read_point_cloud(i)
    pcd = pcd.random_down_sample(sampling_ratio=0.3)
    data = np.asarray(pcd.points)
    rips = Rips()
    dgms = rips.fit_transform(data)
    H0_dgm = dgms[0]
    H1_dgm = dgms[1]
    # print(H0_dgm)
    s0+=np.sum(H0_dgm[1:-1][:,1] - H0_dgm[1:-1][:,0])/len(H0_dgm[1:-1])
    s1+=np.sum(H1_dgm[:,1] - H1_dgm[:,0])/len(H1_dgm)
    print(s0, s1)
avg0 = s0/len(complete[:100])
avg1 = s1/len(complete[:100])
print(avg0, avg1)
```
